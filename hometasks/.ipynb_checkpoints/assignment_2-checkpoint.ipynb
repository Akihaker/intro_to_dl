```python
import torch
from torch import nn
from torch import optim

class MyNet(nn.Module):
    def __init__(self, seed=42):
      
        super().__init__()
        torch.manual_seed(seed)

        self.lstm = nn.LSTM()
        

    def forward(self, user_input, item_input):
        item_are_output = self.item_model(item_input)
        user_are_output = self.user_model(user_input).T
        blanding_output = item_are_output * self.alpha + user_are_output * (1 - self.alpha)

        output, reg = self.autoencoder(blanding_output)

        return output, reg
```
